```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv(r'C:\Users\marci\DataScience\Pitch_predictions\pitch-aileron\train.csv').copy()
print(df.head())
print(df.shape)
```

```python
corr_matrix = df.corr()

corr_matrix['Target'].sort_values(ascending=False)
```

```python
df_T= df.drop('Target', axis=1)
```

```python
df_T.info()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
df_scaled = pd.DataFrame(ss.fit_transform(df_T),columns = df_T.columns)
```

```python
df_scaled
```

```python
df_scaled_r = df_scaled.drop(columns=[ 'SGZ'], inplace=True, axis=1)

df_scaled_r
```

```python
df_scaled
```

```python
df_labels = df['Target'].copy()
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(df_scaled, df_labels)
```

```python
from sklearn.metrics import mean_squared_error
df_pred=lin_reg.predict(df_scaled)
lin_mse = mean_squared_error(df_labels, df_pred)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
df_test=pd.read_csv(r'C:\Users\marci\DataScience\Pitch_predictions\pitch-aileron\test.csv')
```

```python
df_test_no_Id = df_test.loc[:, df_test.columns!='Id'].copy()

df_test_no_Id
```

```python
df_test_scaled = pd.DataFrame(ss.fit_transform(df_test_no_Id),columns = df_test_no_Id.columns).copy()

df_test_scaled.drop(columns=[ 'SGZ'], inplace=True, axis=1)
```

```python
y_hats = lin_reg.predict(df_test_scaled)

df_test['prediction'] = y_hats

df_out = pd.merge(df_test, df_test[['prediction']], how = 'left', left_index = True, right_index = True)
```
